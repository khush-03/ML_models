```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_excel(r'C:\Users\khush\Downloads\test.xlsx')
```

```python
df.isnull()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(scaled_data)
clusters = kmeans.labels_
df['cluster'] = clusters
```

```python
def predict_cluster(data_point):
    scaled_point = scaler.transform([data_point])
    cluster = kmeans.predict(scaled_point)
    return cluster[0]
```

```python
def explain_cluster(data_point):
    scaled_point = scaler.transform([data_point])
    distances = kmeans.transform(scaled_point)
    cluster = kmeans.predict(scaled_point)[0]
    return f"The data point is closest to cluster {cluster} with distances to centroids: {distances[0]}"
```

```python
data_point = df.iloc[0, :-1].values
predicted_cluster = predict_cluster(data_point)
print(f"The new data point belongs to cluster: {predicted_cluster}")
print(explain_cluster(data_point))
```
